```python

##############################
# Diabete Feature Engineering
##############################

# Problem : zellikleri belirtildiinde kiilerin diyabet hastas olup olmadklarn tahmin edebilecek bir makine renmesi modeli gelitirilmesi
# istenmektedir. Modeli gelitirmeden nce gerekli olan veri analizi ve zellik mhendislii admlarn gerekletirmeniz beklenmektedir.

# Veri seti ABD'deki Ulusal Diyabet-Sindirim-Bbrek Hastalklar Enstitleri'nde tutulan byk veri setinin parasdr.
# ABD'deki Arizona Eyaleti'nin en byk 5. ehri olan Phoenix ehrinde yaayan 21 ya ve zerinde olan Pima Indian kadnlar
# zerinde yaplan diyabet aratrmas iin kullanlan verilerdir. 768 gzlem ve 8 saysal bamsz deikenden olumaktadr.
# Hedef deiken "outcome" olarak belirtilmi olup; 1 diyabet test sonucunun pozitif oluunu, 0 ise negatif oluunu belirtmektedir.

# Pregnancies: Hamilelik says
# Glucose: Glikoz
# BloodPressure: Kan basnc (Diastolic(Kk Tansiyon))
# SkinThickness: Cilt Kalnl
# Insulin: nslin.
# BMI: Beden kitle indeksi.
# DiabetesPedigreeFunction: Soyumuzdaki kiilere gre diyabet olma ihtimalimizi hesaplayan bir fonksiyon.
# Age: Ya (yl)
# Outcome: Kiinin diyabet olup olmad bilgisi. Hastala sahip (1) ya da deil (0)


# GREV 1: KEFC VER ANALZ
           # Adm 1: Genel resmi inceleyiniz.
           # Adm 2: Numerik ve kategorik deikenleri yakalaynz.
           # Adm 3:  Numerik ve kategorik deikenlerin analizini yapnz.
           # Adm 4: Hedef deiken analizi yapnz. (Kategorik deikenlere gre hedef deikenin ortalamas, hedef deikene gre numerik deikenlerin ortalamas)
           # Adm 5: Aykr gzlem analizi yapnz.
           # Adm 6: Eksik gzlem analizi yapnz.
           # Adm 7: Korelasyon analizi yapnz.

# GREV 2: FEATURE ENGINEERING
           # Adm 1:  Eksik ve aykr deerler iin gerekli ilemleri yapnz. Veri setinde eksik gzlem bulunmamakta ama Glikoz, Insulin vb.
           # deikenlerde 0 deeri ieren gzlem birimleri eksik deeri ifade ediyor olabilir. rnein; bir kiinin glikoz veya insulin deeri
           # 0 olamayacaktr. Bu durumu dikkate alarak sfr deerlerini ilgili deerlerde NaN olarak atama yapp sonrasnda eksik deerlere
           # ilemleri uygulayabilirsiniz.
           # Adm 2: Yeni deikenler oluturunuz.
           # Adm 3:  Encoding ilemlerini gerekletiriniz.
           # Adm 4: Numerik deikenler iin standartlatrma yapnz.
           # Adm 5: Model oluturunuz.


# Gerekli Ktphane ve Fonksiyonlar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,roc_auc_score
# from sklearn.model_selection import GridSearchCV, cross_validate
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import warnings
warnings.simplefilter(action="ignore")

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
#pd.set_option('display.max_rows', 20)
pd.set_option('display.float_format', lambda x: '%.3f' % x)


df = pd.read_csv("diabetes.csv")
df.head()

##################################
# GREV 1: KEFC VER ANALZ
##################################

##################################
# GENEL RESM
##################################

def check_df(dataframe, head=5):
    print("##################### Shape #####################")
    print(dataframe.shape)
    print("##################### Types #####################")
    print(dataframe.dtypes)
    print("##################### Head #####################")
    print(dataframe.head(head))
    print("##################### Tail #####################")
    print(dataframe.tail(head))
    print("##################### NA #####################")
    print(dataframe.isnull().sum()) #eksik deger var m? varsa kac tane?
    print("##################### Quantiles #####################")
    print(dataframe.quantile([0, 0.05, 0.50, 0.95, 0.99, 1]).T) # saysal deikenlerin ceyrekliklerinin incelenmesi

check_df(df)
# Glucose degeri sfr olabilir mi?
# Insulin degeri sfr olabilir mi ?
# Kan basnc sfr olabilir mi?
# Veri setinde eksik degereler vard da sfr basld?
# Insulin degerinde 95 ceyreklikten max degere buyuk bir scrays var bu da aykr deger olabileceginin bir sinyali.

df.head()

##################################
# NUMERK VE KATEGORK DEKENLERN YAKALANMASI
##################################

def grab_col_names(dataframe, cat_th=10, car_th=20):
    """

    Veri setindeki kategorik, numerik ve kategorik fakat kardinal deikenlerin isimlerini verir.
    Not: Kategorik deikenlerin ierisine numerik grnml kategorik deikenler de dahildir.

    Parameters
    ------
        dataframe: dataframe
                Deiken isimleri alnmak istenilen dataframe
        cat_th: int, optional
                numerik fakat kategorik olan deikenler iin snf eik deeri
        car_th: int, optional
                kategorik fakat kardinal deikenler iin snf eik deeri

    Returns
    ------
        cat_cols: list
                Kategorik deiken listesi
        num_cols: list
                Numerik deiken listesi
        cat_but_car: list
                Kategorik grnml kardinal deiken listesi

    Examples
    ------
        import seaborn as sns
        df = sns.load_dataset("iris")
        print(grab_col_names(df))


    Notes
    ------
        cat_cols + num_cols + cat_but_car = toplam deiken says
        num_but_cat cat_cols'un ierisinde.

    """
    # cat_cols, cat_but_car
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and dataframe[col].dtypes != "O"] # 0,1,2
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and dataframe[col].dtypes == "O"] # name
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car] # cat_cols tm object  veri tipini tuttugu iin ierisinde cat_but_car bulunabilir.

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat] #numerik_gorunumlu kategorikler hari

    print(f"Observations: {dataframe.shape[0]}") # satr
    print(f"Variables: {dataframe.shape[1]}") # sutun
    print(f'cat_cols: {len(cat_cols)}') # categorik degiken says
    print(f'num_cols: {len(num_cols)}') # numerik deikenler
    print(f'cat_but_car: {len(cat_but_car)}') # categorik fakat kardinal
    print(f'num_but_cat: {len(num_but_cat)}') # numerik grnml kategorik

    return cat_cols, num_cols, cat_but_car




cat_cols, num_cols, cat_but_car = grab_col_names(df)



cat_cols
num_cols
cat_but_car


##################################
# KATEGORK DEKENLERN ANALZ
##################################

# amacm deikene dair degerlerin oranna gz atmak.
def cat_summary(dataframe, col_name, plot=True): # plot:true olursa if alr.
    print(pd.DataFrame({col_name: dataframe[col_name].value_counts(),  #deikende hangi degerden kacar adet var?
                        "Ratio": 100 * dataframe[col_name].value_counts() / len(dataframe)})) # deger adetlerini toplam deger saysna blm oran verir.
    print("##########################################")
    if plot:
        sns.countplot(x=dataframe[col_name], data=dataframe)
        plt.show()

# kategorik deikenimde deniyorum.
cat_summary(df, "Outcome")


for col in cat_cols:
    cat_summary(df, col)

##################################
# NUMERK DEKENLERN ANALZ
##################################

def num_summary(dataframe, numerical_col, plot=False):  # plot:true olursa if alr.
    quantiles = [0.05, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90, 0.95, 0.99] # hangi ceyreklikleri istiyorum?
    print(dataframe[numerical_col].describe(quantiles).T) # istedigim ceyreklikler baznda describe gz atyorum.

    if plot:
        dataframe[numerical_col].hist(bins=20)
        plt.xlabel(numerical_col)
        plt.title(numerical_col)
        plt.show()

for col in num_cols: # num_cols: grab_col_names fonksiyonundan elde ettigim numerik deikenlerim.
    num_summary(df, col, plot=False)

##################################
# NUMERK DEKENLERN TARGET GRE ANALZ
##################################

# numerik degikenlerin target deikene gre ortalamalarn inceleyelim:
def target_summary_with_num(dataframe, target, numerical_col):
    print(dataframe.groupby(target).agg({numerical_col: "mean"}), end="\n\n\n")


for col in num_cols:
    target_summary_with_num(df, "Outcome", col)


##################################
# KORELASYON
##################################

# Korelasyon, olaslk kuram ve istatistikte iki rassal deiken arasndaki dorusal ilikinin ynn ve gcn belirtir

df.corr()

# Korelasyon Matrisi
f, ax = plt.subplots(figsize=[18, 13])
sns.heatmap(df.corr(), annot=True, fmt=".2f", ax=ax, cmap="magma")
ax.set_title("Correlation Matrix", fontsize=20)
plt.show()



##################################
# BASE MODEL KURULUMU
##################################
# Bu konular ileride grlecek... O yzden imdilik buray ezbere geelim.
# Amacmz herhangi bir ilem yapmadan baarmz ne durumda?
# Sonrasyla karslastralm.

y = df["Outcome"]
X = df.drop("Outcome", axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=17)

rf_model = RandomForestClassifier(random_state=46).fit(X_train, y_train)
y_pred = rf_model.predict(X_test)


print(f"Accuracy: {round(accuracy_score(y_pred, y_test), 2)}") # basar oran
print(f"Recall: {round(recall_score(y_pred,y_test),3)}") # Gercekte diyabet olanlarn kacna diyabet dedigi
print(f"Precision: {round(precision_score(y_pred,y_test), 2)}") # Recall'in tam tersi. Model tarafndan tahmin edilen degerlerin kac tanesi diyabet
print(f"F1: {round(f1_score(y_pred,y_test), 2)}") # Recall ve precision ortalamas
print(f"Auc: {round(roc_auc_score(y_pred,y_test), 2)}") # farkl snflandrma esik degerlerine gre basar

# Accuracy: 0.77
# Recall: 0.706 # pozitif snfn ne kadar baarl tahmin edildii
# Precision: 0.59 # Pozitif snf olarak tahmin edilen deerlerin baars
# F1: 0.64
# Auc: 0.75


# Model hangi deikene daha cok nem varmi?
def plot_importance(model, features, num=len(X), save=False):
    feature_imp = pd.DataFrame({'Value': model.feature_importances_, 'Feature': features.columns})
    plt.figure(figsize=(10, 10))
    sns.set(font_scale=1)
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value",
                                                                     ascending=False)[0:num])
    plt.title('Features')
    plt.tight_layout()
    plt.show()
    if save:
        plt.savefig('importances.png')

plot_importance(rf_model, X)



##################################
# GREV 2: FEATURE ENGINEERING
##################################

##################################
# EKSK DEER ANALZ
##################################
df.isnull().sum() # eksik degerler yoktu. Fakat sfr olamayacak deikenlere sfr atanmst.
df.describe()
# Bir insanda Pregnancies ve Outcome dndaki deiken deerleri 0 olamayaca bilinmektedir.
# Bundan dolay bu deerlerle ilgili aksiyon karar alnmaldr. 0 olan deerlere NaN atanabilir .

# minimum degeri sfr olamayacak deikenler yakalanyor.
# kategorik deikenler hari braklyor.
zero_columns = [col for col in df.columns if (df[col].min() == 0 and col not in ["Pregnancies", "Outcome"])]

zero_columns

# Gzlem birimlerinde 0 olan degiskenlerin her birisine gidip 0 iceren gozlem degerlerini NaN ile deitirdik.

# where ile eger ki art saglanyorsa NAN yazacagm, saglanmyorsa oldugu gibi yazacagm.
for col in zero_columns:
    df[col] = np.where(df[col] == 0, np.nan, df[col])

# Eksik Gzlem Analizi
df.isnull().sum()

# artk eksik degerleri (NAN) incelebilirz.
def missing_values_table(dataframe, na_name=False):
    na_columns = [col for col in dataframe.columns if dataframe[col].isnull().sum() > 0] # eksik deger varsa na_columns deikeninde tutulur.
    n_miss = dataframe[na_columns].isnull().sum().sort_values(ascending=False) # sralamann sebebi ilk olarak fazla eksik degere sahip eikenleri grmek istememiz.
    ratio = (dataframe[na_columns].isnull().sum() / dataframe.shape[0] * 100).sort_values(ascending=False) # eksik degerlerin tm deerler ierisindeki oran
    missing_df = pd.concat([n_miss, np.round(ratio, 2)], axis=1, keys=['n_miss', 'ratio']) # kac deger var ve orann birletiriyoruz.
    print(missing_df, end="\n")
    if na_name: #na_name true ise degeri dndrr.
        return na_columns

na_columns = missing_values_table(df, na_name=True)


# Eksik Deerlerin Baml Deiken ile likisinin ncelenmesi
# amacmz eksik degerler ile var olan degerlerin karslastrmasn yapmak olacak.

def missing_vs_target(dataframe, target, na_columns):
    temp_df = dataframe.copy()
    for col in na_columns: # eksik degeri olan deikenlerde geziyoruz.
        temp_df[col + '_NA_FLAG'] = np.where(temp_df[col].isnull(), 1, 0) # temp_df[col].isnull() degeri true false olarak dndrr. true ise 1: false ise 0 yazar.
        # bu ilemin amac eksik olan degerlerde var olan degerleri ayrstrmaktr.
    na_flags = temp_df.loc[:, temp_df.columns.str.contains("_NA_")].columns # NA barndran deikenlerde gezmek istiyorum. yeni deikene atadm.
    for col in na_flags:
        print(pd.DataFrame({"TARGET_MEAN": temp_df.groupby(col)[target].mean(),
                            "Count": temp_df.groupby(col)[target].count()}), end="\n\n\n")


missing_vs_target(df, "Outcome", na_columns)
# Eksikse 1 deilse 0.




# Eksik Deerlerin Doldurulmas
for col in zero_columns:
    df.loc[df[col].isnull(), col] = df[col].median()


df.isnull().sum()

df.describe()

##################################
# AYKIRI DEER ANALZ
##################################


# aykr degerler iin limit belirleme
def outlier_thresholds(dataframe, col_name, q1=0.05, q3=0.95):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit


# aykr deger var m yok mu?
def check_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False




# aykr degerleri basklama
def replace_with_thresholds(dataframe, variable, q1=0.05, q3=0.95):
    low_limit, up_limit = outlier_thresholds(dataframe, variable, q1=0.05, q3=0.95)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit


# sns.boxplot(x=df["Insulin"])

# Aykr Deer Analizi ve Basklama lemi
for col in df.columns:
    print(col, check_outlier(df, col))
    if check_outlier(df, col):
        replace_with_thresholds(df, col)


for col in df.columns:
    print(col, check_outlier(df, col))



##################################
# ZELLK IKARIMI
##################################

# Ya deikenini kategorilere ayrp yeni ya deikeni oluturulmas
df.loc[(df["Age"] >= 21) & (df["Age"] < 50), "NEW_AGE_CAT"] = "mature"
df.loc[(df["Age"] >= 50), "NEW_AGE_CAT"] = "senior"


df.head()

# BMI 18,5 aas underweight, 18.5 ile 24.9 aras normal, 24.9 ile 29.9 aras Overweight ve 30 st obez
df['NEW_BMI'] = pd.cut(x=df['BMI'], bins=[0, 18.5, 24.9, 29.9, 100],labels=["Underweight", "Healthy", "Overweight", "Obese"])
#literatur taramasndan elde ettigimiz degerler.

# Glukoz degerini kategorik deikene evirme
df["NEW_GLUCOSE"] = pd.cut(x=df["Glucose"], bins=[0, 140, 200, 300], labels=["Normal", "Prediabetes", "Diabetes"])

# # Ya ve beden kitle indeksini bir arada dnerek kategorik deiken oluturma 3 krlm yakaland
df.loc[(df["BMI"] < 18.5) & ((df["Age"] >= 21) & (df["Age"] < 50)), "NEW_AGE_BMI_NOM"] = "underweightmature"
df.loc[(df["BMI"] < 18.5) & (df["Age"] >= 50), "NEW_AGE_BMI_NOM"] = "underweightsenior"
df.loc[((df["BMI"] >= 18.5) & (df["BMI"] < 25)) & ((df["Age"] >= 21) & (df["Age"] < 50)), "NEW_AGE_BMI_NOM"] = "healthymature"
df.loc[((df["BMI"] >= 18.5) & (df["BMI"] < 25)) & (df["Age"] >= 50), "NEW_AGE_BMI_NOM"] = "healthysenior"
df.loc[((df["BMI"] >= 25) & (df["BMI"] < 30)) & ((df["Age"] >= 21) & (df["Age"] < 50)), "NEW_AGE_BMI_NOM"] = "overweightmature"
df.loc[((df["BMI"] >= 25) & (df["BMI"] < 30)) & (df["Age"] >= 50), "NEW_AGE_BMI_NOM"] = "overweightsenior"
df.loc[(df["BMI"] > 18.5) & ((df["Age"] >= 21) & (df["Age"] < 50)), "NEW_AGE_BMI_NOM"] = "obesemature"
df.loc[(df["BMI"] > 18.5) & (df["Age"] >= 50), "NEW_AGE_BMI_NOM"] = "obesesenior"

# Ya ve Glikoz deerlerini bir arada dnerek kategorik deiken oluturma
df.loc[(df["Glucose"] < 70) & ((df["Age"] >= 21) & (df["Age"] < 50)), "NEW_AGE_GLUCOSE_NOM"] = "lowmature"
df.loc[(df["Glucose"] < 70) & (df["Age"] >= 50), "NEW_AGE_GLUCOSE_NOM"] = "lowsenior"
df.loc[((df["Glucose"] >= 70) & (df["Glucose"] < 100)) & ((df["Age"] >= 21) & (df["Age"] < 50)), "NEW_AGE_GLUCOSE_NOM"] = "normalmature"
df.loc[((df["Glucose"] >= 70) & (df["Glucose"] < 100)) & (df["Age"] >= 50), "NEW_AGE_GLUCOSE_NOM"] = "normalsenior"
df.loc[((df["Glucose"] >= 100) & (df["Glucose"] <= 125)) & ((df["Age"] >= 21) & (df["Age"] < 50)), "NEW_AGE_GLUCOSE_NOM"] = "hiddenmature"
df.loc[((df["Glucose"] >= 100) & (df["Glucose"] <= 125)) & (df["Age"] >= 50), "NEW_AGE_GLUCOSE_NOM"] = "hiddensenior"
df.loc[(df["Glucose"] > 125) & ((df["Age"] >= 21) & (df["Age"] < 50)), "NEW_AGE_GLUCOSE_NOM"] = "highmature"
df.loc[(df["Glucose"] > 125) & (df["Age"] >= 50), "NEW_AGE_GLUCOSE_NOM"] = "highsenior"


# nsulin Deeri ile Kategorik deiken tretmek
def set_insulin(dataframe, col_name="Insulin"):
    if 16 <= dataframe[col_name] <= 166:
        return "Normal"
    else:
        return "Abnormal"


df["NEW_INSULIN_SCORE"] = df.apply(set_insulin, axis=1)

df.head()

df["NEW_GLUCOSE*INSULIN"] = df["Glucose"] * df["Insulin"]

# sfr olan deerler dikkat!!!!
df["NEW_GLUCOSE*PREGNANCIES"] = df["Glucose"] * df["Pregnancies"]
#df["NEW_GLUCOSE*PREGNANCIES"] = df["Glucose"] * (1+ df["Pregnancies"])

df.head()

# Kolonlarn byltlmesi
df.columns = [col.upper() for col in df.columns]

df.head()

##################################
# ENCODING
##################################
num_cols
# Deikenlerin tiplerine gre ayrlmas ilemi
cat_cols, num_cols, cat_but_car = grab_col_names(df)

# LABEL ENCODING
def label_encoder(dataframe, binary_col):
    labelencoder = LabelEncoder()
    dataframe[binary_col] = labelencoder.fit_transform(dataframe[binary_col])
    return dataframe

binary_cols = [col for col in df.columns if df[col].dtypes == "O" and df[col].nunique() == 2]
binary_cols

for col in binary_cols:
    df = label_encoder(df, col)

df.head()

# One-Hot Encoding lemi
# cat_cols listesinin gncelleme ilemi
# target deikenimi ckaryorum.
# bir de binary_cols, zaten daha ncesinde label encoder uygulamstm.
cat_cols = [col for col in cat_cols if col not in binary_cols and col not in ["OUTCOME"]]
cat_cols

def one_hot_encoder(dataframe, categorical_cols, drop_first=False):
    dataframe = pd.get_dummies(dataframe, columns=categorical_cols, drop_first=drop_first)
    return dataframe

df = one_hot_encoder(df, cat_cols, drop_first=True)


df.head()

##################################
# STANDARTLATIRMA
##################################

num_cols

scaler = StandardScaler() # ortalamas sfr, standart sapmas bir olacak sekilde standardize ediyor.
df[num_cols] = scaler.fit_transform(df[num_cols])

df.head()
df.shape

df.describe()

df.shape

##################################
# MODELLEME
##################################

# Feature Engineering ardndan model basarsn degerlendirelim.

df.head()
df.columns


```
